```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Dataset\healthcare_dataset.csv')
```

```python
df.info()
```

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'],format='%d-%m-%y')
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'],format='%d-%m-%y')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
Total_Patients = df.shape[0]
print("Total Patients:", Total_Patients)
```

```python
gender_counts = df['Gender'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Male and Female Distribution')
plt.axis('equal')
plt.show()
```

```python
medical_condition_counts = df['Medical Condition'].value_counts()

top_3_medical_conditions = medical_condition_counts.nlargest(3)

# Plotting a pie chart
plt.figure(figsize=(8, 6))
plt.pie(top_3_medical_conditions, labels=top_3_medical_conditions.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 3 Medical Conditions')
plt.axis('equal')  
plt.show()
```

```python
admission_type_counts = df['Admission Type'].value_counts()

top_3_admission_types = admission_type_counts.nlargest(3)

# Plotting a pie chart
plt.figure(figsize=(8, 6))
plt.pie(top_3_admission_types, labels=top_3_admission_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 3 Admission Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
blood_type_distribution = df['Blood Type'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(8, 6))
plt.pie(blood_type_distribution, labels=blood_type_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Blood Type Distribution')
plt.axis('equal') 
plt.show()
```

```python
patient_counts = df.groupby('year of admission').size()
average_patients_per_year = patient_counts.mean()
print("Average Patients Per Year:", average_patients_per_year)
```

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
max_date_of_admission = df['Date of Admission'].max()
filtered_data = df[df['Date of Admission'] <= max_date_of_admission]
filtered_data = filtered_data[filtered_data['Date of Admission'].dt.year == max_date_of_admission.year]
result = len(filtered_data)
print("Count:", result)
```

```python
total_doctors = df['Doctor'].nunique()

print("Total Doctors:", total_doctors)
```

```python
total_hospitals = df['Hospital'].nunique()

print("Total Hospitals:", total_hospitals)
```

```python
total_count = len(df)

age_group_counts = df['age group'].value_counts()
percentage_per_age_group = (age_group_counts * 100.0) / total_count

result = pd.DataFrame({'age group': age_group_counts.index, 'Percentage': percentage_per_age_group}).reset_index(drop=True)

# Plotting a pie chart
plt.figure(figsize=(8, 6))
plt.pie(result['Percentage'], labels=result['age group'], autopct='%1.1f%%', startangle=140)
plt.title('Age Group Distribution')
plt.axis('equal')  
plt.show()
```
